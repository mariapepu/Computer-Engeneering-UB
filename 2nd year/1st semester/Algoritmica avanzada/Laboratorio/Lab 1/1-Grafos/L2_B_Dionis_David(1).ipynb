<h1>Problema 1: Dijkstra</h1>
<p>
    Implementeu l'algorisme <b>Dijkstra</b>. Donat un graf de la llibreria networX, la funció <b>dijkstra()</b> ha de retornar, si existeix, el camí mínim entre dos nodes. <br>
    A la capçalera del mètode s'expliquen els paràmetres d'entrada i els paràmetres de sortida de la funció. <br>
    Recordeu que és necessari <b>no canviar l'estructura de la funció</b>. No canvieu ni el nom de la funció, ni el nom dels paràmetres d'entrada ni els de sortida. <br>
    - <b>Quina complexitat té aquest algorisme?</b>
</p>    
    
</div>

````python
import heapq

# En Python, podeu representar l'infinit com a float('inf')
def dijkstra(G, origin, destination):

    def backtrace(prev, start, end):
        path = [end]
        while path[-1] != start:
            path.append(prev[path[-1]])
        return path[::-1]

    previous = {}

    # Inicialitzem els nodes com a no visitats i el cost de cada node a l'infinit.
    distances = {v: float('inf') for v in list(nx.nodes(G))} 

    # Inicialitzem un array de visitats
    visited = set()

    # Inicialitzem una cua de prioritat com buida
    hp = []

    # Inicialitzem el cost de l'origen a 0
    distances[origin] = 0

    # Afegim l'origen a la cua de prioritat
    heapq.heappush(hp, (distances[origin], origin))

    # Inicialitzem els nodes total visitats com 0
    expanded = 0

    while hp.__len__() > 0:
        # Obtenim el primer element de la cua de prioritat
        current = heapq.heappop(hp)[1]

        # Si el node actual ja està visitat, passem al següent
        if current in visited:
            break
        
        # Si no, l'afegim a la llista de visitats
        visited.add(current)

        # Recorrem tots els nodes veïns de l'actual
        for neighbor in nx.neighbors(G, current):
            # Si el veïn no està visitat sumem 1 als nodes total visitats 
            if neighbor not in visited:
                expanded +=1

            # Calculem el cost de la nova distància
            new_distance = distances[current] + 1
            
            # si el cost de la nova distancia es menor que la del veïn,
            # actualitzem la distancia i el anterior. Tambe l'agefim a
            # la cua de prioritat
            if new_distance < distances[neighbor]:
                distances[neighbor] = new_distance
                previous[neighbor] = current
                heapq.heappush(hp, (distances[neighbor], neighbor))

    # fem un backtrace per acabar de calcular el camí mínim
    path = backtrace(previous, origin, destination)
    
    return {
        'path': path,
        'expanded': expanded,
        'distance': path.__len__()
    }
````

````python
# Proveu el vostre algorisme!

# Creem un laberint i n'extraiem el graph
maze = Maze(30, 10, seed=17)
G = maze.maze_graph()

# Punt inici i punt final
A = 0
B = len(G.nodes)-1

# Cridem a la punció que heu creat
values = dijkstra(G, A, B)

print('La distància del camí més curt és:', values['distance'])
print('Els nodes que hem expandit són:', values['expanded'])

# Mostrem el camí
maze.set_path(values['path'])
maze.display_maze(plot_path=True, height=200)
````

<h1>Problema 2: Checkpoint</h1>
<p>
    Implementeu una funció anomenada <b>checkpoint()</b> que, donat un graf, un node origen, un node destí i un node extra, trobi el camí més curt des del node origen fins el node destí passant pel node extra.<br>
    - <b>Quina és la complexitat d'aquest algorisme?</b><br>
</p>    
    
</div>

````python
# Aquesta funcion simplement uneix dos dijkstra amb un i elimina 
# el ultim element del path 1 perque no es repeteixi amb el primer 
# del path 2
def checkpoint(G, origin, destination, extra):
    path1 = dijkstra(G, origin, extra)
    path1['path'].pop() # Eliminem el punt extra
    
    path2 = dijkstra(G, extra, destination)
    path = path1['path'] + path2['path']
    distance = path.__len__()
    return {
        'path': path, 
        'distance': distance
    }
````

````python

# Proveu el vostre algorisme!

# Creem un laberint i n'extraiem el graph
maze = Maze(30, 10, seed=17)
G = maze.maze_graph()

# Definim l'inici, el fi i un node extra
A = 0
B = len(G.nodes)-1
E = 13

# La funció set_value ens permet 'activar' un node per a que surti dibuixat al laberint.
maze.set_value(E, 1)

values = checkpoint(G, A, B, E)
print('La distàcia del camí més curt que passa per E és:', values['distance'])
maze.set_path(values['path'])
maze.display_maze(plot_path=True, height=200)

# De la mateixa forma la funció set_value també ens permet 'desactivar' els nodes.
maze.set_value(E, 0)
````

<div class="alert alert-success">
<h1>Problema 3: Checkpoints list</h1>
<p>
    Implementeu una funció anomenada <b>checkpoints_list()</b> que, donat un graf, un node origen, un node destí i una llista de nodes extres, trobi el camí més curt des del node origen fins el node destí passant per tots els nodes extra, en qualsevol ordre.<br>
    - <b>Quina és la complexitat d'aquest algorisme?</b><br>
</p>    
    
</div>

````python
import itertools as it

# Aquest checkpoint permet trobar el cami mes curt d'un origin
# a un destí amb una llista de punts intermigos. Aquest punts intermigos
# no tenen que respectar un ordre.
# Per aixó fem us de les permutacions per trobar tots les possibles
# combinacions. Calcularem la distància de cada combinació i guardarem
# la mes curta junt el seu cami.
def checkpoints_list(G, origin, destination, extras):
    min_path = []

    min_distance = float('inf')

    extras = sorted(extras)

    permutations = list(it.permutations(extras))

    for permutation in permutations:
        temp_path = []

        prev = permutation[0]
        for i in range(1, len(permutation)):
            values = dijkstra(G, prev, permutation[i])
            prev = permutation[i]

            values['path'].pop()
            temp_path += values['path']

        values = dijkstra(G, origin, temp_path[0])
        values['path'].pop()
        temp_path = values['path'] + temp_path

        values = dijkstra(G, temp_path[-1], destination)
        temp_path.pop()
        temp_path += values['path']

        if temp_path.__len__() < min_distance:
            min_distance = temp_path.__len__()
            min_path = temp_path
    
    return {
        'path': min_path, 
        'distance': min_path.__len__()
    }
````

````python
# Proveu el vostre algorisme!

# Creem un laberint i n'extraiem el graph
maze = Maze(30, 10, seed=17)
G = maze.maze_graph()

# Definim l'inici, el fi i node extres
A = 0
B = len(G.nodes)-1
E = [286, 201, 16, 68]

# La funció set_dijk ens permet 'activar' nodes per a que surtin dibuixat al laberint.
maze.set_values(E, 1)

values = checkpoints_list(G, A, B, E)
print('La distàcia del camí més curt que passa per E és:', values['distance'])

maze.set_path(values['path'])
maze.display_maze(plot_path=True, height=200)
````

<h1>Problema 4: Holes</h1>
<p>
    Implementeu una funció anomenada <b>holes()</b> que, donat un graf, un node orígen, un node destí, una llista de nodes i un paràmetre enter, trobi el camí més curt entre orígen i destí.<br>
    Si el camí passa per un dels nodes de la llista, se li aplicarà una penalització.<br>
    - <b>Quina complexitat té aquest algorisme?</b>
</p>    
    
</div>

````python
# Mateix codí que el de dijkstra excepte la part comentada 
def holes(G, origin, destination, holes_list=[], penalty=50): 
    def backtrace(prev, start, end):
        path = [end]
        while path[-1] != start:
            path.append(prev[path[-1]])
        return path[::-1]

    previous = {}

    distances = {v: float('inf') for v in list(nx.nodes(G))} 

    visited = set()

    hp = []

    distances[origin] = 0

    heapq.heappush(hp, (distances[origin], origin))

    expanded = 0

    while hp.__len__() > 0:
        current = heapq.heappop(hp)[1]

        if current in visited:
            break

        visited.add(current)

        for neighbor in nx.neighbors(G, current):
            if neighbor not in visited:
                expanded +=1

            # Aquesta part es l'unica que varia amb dijkstra
            # i es la que ens permet afegi penalitzacions
            # si els nodes estan a la llista de penalitzats (holes_list)
            if neighbor in holes_list:
                new_distance = distances[current] + penalty +1
            else:
                new_distance = distances[current] + 1
                
            if new_distance < distances[neighbor]:
                distances[neighbor] = new_distance
                previous[neighbor] = current
                heapq.heappush(hp, (distances[neighbor], neighbor))
  
    path = backtrace(previous, origin, destination)
    
    return {
        'path': path,
        'expanded': expanded,
        'distance': path.__len__()
    }
    
````

````python
# Proveu el vostre algorisme!
A = 0
B = len(G.nodes)-1
holes_list = [4, 190, 265]
color = (255, 0, 255)

penalties = [2,5,10]
for v in penalties:
    maze.set_values(holes_list, 1, color)
    values = holes(G, A, B, holes_list, v)
    
    maze.set_path(values['path'])
    print('Penalització:',v)
    print('Cost:',values['distance'])
    maze.display_maze(plot_path=True, height=200)

    maze.set_values(holes_list, 0, color)
````

<h1>Problema 5: Money</h1>
<p>
    Afegirem recompenses (valors negatius) al passar per determinades cel·les del laberint.<br>
    - <b>Podem aplicar Dijkstra en aquesta situació?</b> <br><br>
    Suposem un cas especial de laberint: Només podem avançar cap a la dreta (est) o cap a baix (sud). Així, el nostre laberint es pot reconvertir en un graf dirigit ja que els camins nomes permeten una direcció.<br>
    - <b>En aquesta nova suposició, quin algorisme ens permet trobar el camí més curt?</b> <br>
    - <b>Per què hem hagut de limitar les direccions permeses?</b><br><br>
    Adapta l'algorisme que ens permet trobar el camí més curt (tenint en compte les recompenses) a aquest problema en concret. La funció, anomenada <b>money()</b> rep com a paràmetres un node origen, un node destí i un diccionari $\{n: w\}$ on $n$ és l'índex d'un node i $w$ és el valor de recompensa del node.
</p>    
    
</div>

````python
def bellman_ford(G, origin, destination, nodes_cost):
    dist = {}
    prev = {}

    # incialitzem totes les distancies
    # amb infinit i totes les distancies
    # al anterior amb None, donat que
    # no es ningu de moment
    for n in G.nodes():
        dist[n] = float('inf')
        prev[n] = None

    # la distancia al origin es de cero
    # donat que partim d'aqui
    dist[origin] = 0

    # iterem tantes vegades com nodes hi han
    for _ in G.nodes():
        # iterem un altre cop donat que desde
        # cada node intentarem arribar a tots 
        # el nodes de manera "bruta"
        for node in G.nodes():
            # estudiarem tots els seus veins
            # donat que son els unics nodes on
            # ens podem moure
            for neighbor in G.neighbors(node):
                # el cost per default es 1
                node_cost = 1

                # si aquest node te penalitzacio,
                # cambiem el cost per aquest valor
                if node in nodes_cost:
                    node_cost = nodes_cost[node]
                
                # si la distancia al vei es més gran que la 
                # distancia al node (del origen al node actual) i al 
                # cost del node actual al vei, cambie els valors de la
                # distancia del origin al vei i diem que l'anterior
                # al vei es el node actual
                if dist[neighbor] > dist[node] + node_cost:
                    dist[neighbor] = dist[node] + node_cost
                    prev[neighbor] = node

    # Let's backtrace
    path = [destination]
    # mentres l'ultim element no sigui l'origin
    while path[-1] != origin:
        # unim el previ de l'ultim
        path.append(prev[path[-1]])

    path = path[::-1]
    # el valor de la distancia al destination
    # sera la gaurdada en path
    return path, dist[destination]
````

````python
# Proveu el vostre algorisme!

# Creem un laberint i n'extraiem el graph
maze = Maze(15, 15, p_down=0.4, p_build=0, directed=True, seed=1)
G = maze.maze_graph()

# Punt inici i punt final
A = 0
B = len(G.nodes)-1
money_dict = {63: -10, 127: -10, 107:-100}
# money_dict = {}

for k,v in money_dict.items():
    maze.set_value(k, v, color=(255,255,0))

# Cridem a la punció que heu creat
path, distance = bellman_ford(G, A, B, money_dict)
print('La distància del camí més curt és:', distance)

# Mostrem el camí
maze.set_path(path)
maze.display_maze(plot_path=True, height=300)

# values = dijkstra(G, A, B)
# print('La distància del camí més curt és:', values['distance'])

# # Mostrem el camí
# maze.set_path(values['path'])
# maze.display_maze(plot_path=True, height=300)
````
