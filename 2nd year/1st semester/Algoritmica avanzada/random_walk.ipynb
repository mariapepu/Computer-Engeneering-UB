```python
import random

def random_walk(G, origen, destino, repeat = False):
    
    # -----------------------------------------------------------------------------

    path = []
    expanded = 0
    
    nodoActual = origen
    
    # -----------------------------------------------------------------------------

    while (nodoActual != destino):
        
        expanded += 1
        path.append(nodoActual)
        
        if (repeat):
            vecinosDisponibles = [nodoVecino
                                  for nodoVecino in G.neighbors(nodoActual)]
        
        else:
            vecinosDisponibles = [nodoVecino
                                  for nodoVecino in G.neighbors(nodoActual)
                                  if nodoVecino not in path]
        
        if (len(vecinosDisponibles) == 0):
            return "No se ha encontrado un camino"
        
        nodoActual = random.choice(vecinosDisponibles)
        
    path.append(destino)
    expanded += 1
    
    # -----------------------------------------------------------------------------
    
    return {'path': path, 'expanded': expanded}
```

```python
G = nx.generators.barabasi_albert_graph(10, 2)
nx.draw_circular(G, with_labels = True)
random_walk(G, 5, 8)
```
