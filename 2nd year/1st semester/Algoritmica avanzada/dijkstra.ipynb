```python
def dijkstra(G, origen, destino, infinity=99999999999):
    
    if origen in G._nodes and destino in G._nodes:
        path = [(0,origen)] #Creamos una lista de tuplas con ('distance',nodo) para acceder rapidamente al de min. dist
        path_min = []
        expanded = 0

        for i in G.nodes():
            G.node[i].update({'distance':infinity,'prev':None,'visit':False})

        G.node[origen]['distance'] = 0

        while len(path) != 0:
            aux= min(path)
            nodo = aux[1]
            
            if nodo==destino:
                break
                
            if G.node[nodo]['visit'] == False:
                expanded += 1
                G.node[nodo]['visit'] = True
                for i in G.neighbors(nodo):
                    if (G.node[i]['distance'] > G.node[nodo]['distance'] + G.edge[nodo][i]['distance']):
                        G.node[i]['distance'] = G.node[nodo]['distance'] + G.edge[nodo][i]['distance']
                        G.node[i]['prev'] = nodo
                        path.append((G._nodes[i]['distance'],i))   
            
            path.remove(aux)

        nodo = destino
        while G.node[nodo]['prev'] != None:
            path_min.append(nodo)
            nodo = G.node[nodo]['prev']
            
        path_min.append(origen)
        path_min.reverse()

        return {
            'path': path_min,    #camino a seguir
            'expanded': expanded, #numero de nodos abiertos
            'distance': G.node[destino]['distance']
        }
    
    else:
        return ("Nodes don't exist in this graf")
```
